```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('avitoo.csv')
```

```python
print("Nombre de lignes :", df.shape[0])
print("Nombre de colonnes :", df.shape[1])
```

```python
df.dtypes
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Prix'], bins=50, kde=True)
plt.title("Distribution du prix des véhicules")
plt.xlabel("Prix (MAD)")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Année-Modèle'], y=df['Prix'])
plt.xticks(rotation=90)
plt.title("Prix en fonction de l’année-modèle")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Kilométrage'], bins=50, kde=True)
plt.title("Distribution du kilométrage")
plt.xlabel("Kilométrage")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df['Kilométrage'],
    y=df['Prix'],
    alpha=0.4
)
plt.title("Relation entre le prix et le kilométrage")
plt.xlabel("Kilométrage")
plt.ylabel("Prix")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Type de carburant')
plt.title("Répartition par type de carburant")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Boite de vitesses')
plt.title("Répartition des boîtes de vitesses")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='État')
plt.title("État des véhicules")
plt.xticks(rotation=45)
plt.show()
```
